```python
X = TSNE(n_components = 3, random_state = 1).fit_transform(df)
xyz = pd.DataFrame(X)
x = xyz[0]
b = xyz[1]
z = xyz[2]

ax = plt.axes(projection='3d')
sets = list(set(y))
for i in range(len(y)):
    y[i] = sets.index(y[i])

fig  = ax.scatter3D(x, b, z, c = y, cmap = 'viridis')

plt.savefig('teste_3d.png', dpi = 300)

```
